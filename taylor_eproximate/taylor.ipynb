# taylor aproximate rule
based on this rule when we have a none linear but derivable function we can aproximate an afine fuction that represent the none linear one

the rule: (f^ = linear fuvtion, x= predicted value, z= real value)
## f^(x) = f(z) + gradian(f(z)).T * (x - z)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# first we make a nonlinear but deriable funtion for example

def f(x):
    return x**3

def f_rad(x): #deivative of the main function
    return 3*x**2

z = 1 # x that will be without error 
x = np.linspace(-2, 3, 400)
y = f(x)
y_rad = f_rad(x)

f_at_z = f(z)
grad_at_z = f_rad(z)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='f(x)')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Nonlinear Polynomial Function')
plt.show()
```

```python
# based on the rule:
def f_hat(x=x, y=y, y_rad=y_rad):
    return f_at_z + grad_at_z * (x - z)

y_pred = f_hat()
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print(f"R²: {r2_score(y, y_pred)}")
print(f"MSE: {mean_squared_error(y, y_pred)}")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='f(x)')
plt.plot(x, y_pred, label="f_hat(x)")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Nonlinear Polynomial Function')
plt.show()
```
